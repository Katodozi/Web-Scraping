```python
#finding elements in web page (by class , id , etc)
web1 = requests.get("https://www.onlinekhabar.com/business")
print(web1)
```

```python
web1.content
soup1 = BeautifulSoup(web1.content, "html.parser")
print(soup1.prettify())
```

```python
tag = soup1.p.string
tag
```

```python
soup1.find_all("p")
```

```python
class_data = soup1.find("div", class_="ok-news-post ok-post-ltr")
class_data
```

```python
class_data.find_all("img")
```

```python
id_data = soup1.find("li",id = "menu-item-735066")
id_data
```

```python
id_data.find("a")
```

```python
#now finally let's learn how to extract text from the tags in the web page 
lines = soup1.find_all("p")
lines
```

```python
for l in lines:
    print(l.text)
```

```python
#extracting links in the web page 
anchor_tags = soup1.find_all("a")
anchor_tags
```

```python
for i in anchor_tags: 
    print(i.get("href")) 
```

```python
#extracting image data from the web page 
for j in soup1.find_all("img"):
    print(j.get("src"))
    #finding it's type 
    print(type(j.get("src")))
    
```
