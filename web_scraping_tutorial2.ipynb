```python
#first , scraping the main page to get the article links
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.financialnotices.com/latest-news.html")
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
#finding the news headlines assuming that they are in an anchor tags
headlines = soup.find_all('div', class_ = "newsitem")
headlines
```

```python
 #Iterate through each headline link
for headline in headlines:
    #get the anchor tag
    anchor = headline.find('a')
    
    #Check if 'href' attribute exists
    if anchor and 'href' in anchor.attrs:
        article_url = anchor['href']  # Get the URL of the article
        
        #the URL is relative, prepend the base URL
        if not article_url.startswith("http"):
            article_url = f"https://www.financialnotices.com/latest-news.html/{article_url}"
        
        # Requesting the article page
        article_response = requests.get(article_url)
        article_soup = BeautifulSoup(article_response.content, 'html.parser')

        #Specifically find paragraphs of the news 
        paragraphs = article_soup.find_all('p')  # Assuming paragraphs are within <p> tags

        # Print out each paragraph's text
        for paragraph in paragraphs:
            print(paragraph.text.strip())  # Print cleaned text of each paragraph
```
